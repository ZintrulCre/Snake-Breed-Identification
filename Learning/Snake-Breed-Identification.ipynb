# 1. Preprocessing

## 1) Input Preprocessing

```python
img_size = 224 # image size
```

```python
images = [] # images
labels = [] # labels matrix
breeds = [] # breeds

label = 0
data_path = "../Data/"
breed_names = os.listdir(data_path)
for breed_name in tqdm(breed_names):
    if breed_name == ".DS_Store":
        continue
    breeds.append(breed_name)
    folder = breed_name + '/'
    img_files = os.listdir(data_path + folder)
    for img_file in img_files:
        try:
            image = cv2.imread(data_path + folder + img_file) # read image
            image = cv2.resize(image, (img_size, img_size)) # resize image
            images.append(image) # append image to list images
            labels.append(label) # append label to list labels
        except Exception as e:
            pass
    label += 1

num_class = label
images = np.asarray(images) # list to np.ndarray
labels = np.asarray(labels) # list to np.ndarray
labels = to_categorical(labels, num_classes=num_class, dtype='int8') # one hot encoding
```

```python
print("images matrix shape: {}".format(images.shape))
print("labels matrix shape: {}".format(labels.shape))
print("breed names: {}".format(breeds))
```

### Image Normalization

#### Plot Original Images

```python
def PlotImages(X):
    fig, axes = plt.subplots(3, 3, figsize=(8, 8))
    axes = axes.flatten()
    for i in tqdm(range(9)):
        axes[i].imshow(X[i])
        axes[i].set_xticks(())
        axes[i].set_yticks(())
    plt.tight_layout()
```

```python
PlotImages(images)
```

#### Normalize Images

```python
def NormalizeImages(images):
    if images.ndim < 3 or images.ndim > 4:
        return None
    if images.ndim == 3:
        images = np.expand_dims(images)
    images = images.astype('float64') # int8 to float64
    for i in tqdm(range(len(images))): # normalization
    #     mean = images[i].mean()
    #     std = images[i].std()
    #     images[i] = (images[i] - mean) / std
        for channel in range(3):
            images[i,:,:,channel] = scale(images[i,:,:,channel])
    return images
```

```python
images = NormalizeImages(images)
```

#### Plot Normalized Images

```python
PlotImages(images)
```

## 2) Callbacks

```python
# callback function for plotting loss 
class History(Callback):
    def on_train_begin(self, logs={}):
        self.losses = {'batch':[], 'epoch':[]}
        self.accuracy = {'batch':[], 'epoch':[]}
        self.val_loss = {'batch':[], 'epoch':[]}
        self.val_acc = {'batch':[], 'epoch':[]}

    def on_batch_end(self, batch, logs={}):
        self.losses['batch'].append(logs.get('loss'))
        self.accuracy['batch'].append(logs.get('acc'))
        self.val_loss['batch'].append(logs.get('val_loss'))
        self.val_acc['batch'].append(logs.get('val_acc'))

    def on_epoch_end(self, batch, logs={}):
        self.losses['epoch'].append(logs.get('loss'))
        self.accuracy['epoch'].append(logs.get('acc'))
        self.val_loss['epoch'].append(logs.get('val_loss'))
        self.val_acc['epoch'].append(logs.get('val_acc'))

    def loss_plot(self, loss_type):
        iters = range(len(self.losses[loss_type]))
        plt.figure()
        # acc
        plt.plot(iters, self.accuracy[loss_type], '#9933ff', label='train accuracy')
        # loss
        plt.plot(iters, self.losses[loss_type], '#1404FB', label='train loss')
        if loss_type == 'epoch':
            # val_acc
            plt.plot(iters, self.val_acc[loss_type], '#ff0d00', label='validation accuracy')
            # val_loss
            plt.plot(iters, self.val_loss[loss_type], '#ff9a00', label='validation loss')
        plt.grid(True)
        plt.xlabel(loss_type)
        plt.ylabel('accuracy and loss')
        plt.legend(loc="upper right")
        plt.show()
```

```python
class ReduceLearningRate(Callback):
    def __init__(self, lr=1e-3, min_lr=1e-9, reduction_rate=0.1, epoch_batch=20):
        self.lr = lr
        self.min_lr = min_lr
        self.reduction_rate = reduction_rate
        self.epoch_batch = epoch_batch
        self.epoch = 0
        
    def on_epoch_end(self, batch, logs={}):
        self.epoch += 1
        if self.epoch == self.epoch_batch:
            if self.lr > self.min_lr:
                self.epoch = 0
                self.lr *= self.reduction_rate
                logs['lr'] = self.lr
            print('\n\nLearning Rate reduced to {}\n'.format(self.lr))
```

```python
# callbacks

# check point
filepath = 'Snake-Breed-Identification.h5'
model_check_point = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True)

# tensorboard
tensor_board = TensorBoard(log_dir='Snake-Breed-Identification', write_images=1, histogram_freq=1)  

# learning rate reduction
learning_rate_reduction = ReduceLROnPlateau(monitor='val_acc', patience=5, verbose=1, factor=0.5, min_lr=1e-9)

# plot loss
history = History()

# callbacks = [model_check_point, tensor_board, learning_rate_reduction, loss_history, reduce_learning_rate]
callbacks = [model_check_point, learning_rate_reduction, tensor_board]
```

# 2. Training

## 1) Build the Model

### Load Model (Optional)

```python
# model = load_model('Snake-Breed-Identification.h5')
```

### Build Model

```python
# read pretrained model, download weights from ImageNet
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_size, img_size, 3))

# freeze layers
for layer in base_model.layers:
    layer.trainable = False

# add top and prediction layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
# x = Dropout(0.05)(x)
x = Dense(16, activation='relu')(x)
x = Dropout(0.05)(x)
predictions = Dense(num_class, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)


model.summary()
```

### Hyperparameters

```python
batch_size = 4
epochs = 1000
learing_rate = 1e-3
```

```python
model.compile(optimizer=Adam(lr=learing_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

## 2) Train the Model

```python
history = model.fit(x_train, y_train, batch_size=batch_size,
                    epochs=epochs, verbose=1,
                    validation_data = (x_val, y_val),
                    callbacks=callbacks)
```

### Training With Generator

```python
history = model.fit_generator(datagen.flow(x_train, y_train, batch_size=batch_size),
                              steps_per_epoch=len(x_train) // batch_size,
                              epochs = epochs, verbose=1,
                              validation_data = (x_val, y_val),
                               callbacks=callbacks)
```

## 3) Evaluate the Model

```python
result = model.evaluate(x_val, y_val)
print("loss: {}".format(result[0]))
print("accuracy: {}".format(result[1]))
```

# 3. Predicting

```python
def Predict(images):
    if images.ndim < 3 or images.ndim > 4:
        return None
    if images.ndim == 3:
        images = np.expand_dims(images, axis=0)
    predictions = model.predict(images)
    for prediction in tqdm(predictions):
        confidence = np.max(prediction)
        breed = breeds[np.argmax(prediction)]
        print("breed: {}\nconfidence: {}\n".format(breed, confidence))
```
