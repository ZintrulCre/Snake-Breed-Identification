```python
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('acc_no_dropout.txt', 'rb') as f:
    acc_no_dropout = pickle.load(f)
with open('acc_no_dropout_validation.txt', 'rb') as f:
    acc_no_dropout_validation = pickle.load(f)
    
with open('acc_dropout_005.txt', 'rb') as f:
    acc_dropout_005 = pickle.load(f)
with open('acc_dropout_005_validation.txt', 'rb') as f:
    acc_dropout_005_validation = pickle.load(f)
    
with open('acc_dropout_010.txt', 'rb') as f:
    acc_dropout_010 = pickle.load(f)
with open('acc_dropout_010_validation.txt', 'rb') as f:
    acc_dropout_010_validation = pickle.load(f)
    
with open('acc_dropout_015.txt', 'rb') as f:
    acc_dropout_015 = pickle.load(f)
with open('acc_dropout_015_validation.txt', 'rb') as f:
    acc_dropout_015_validation = pickle.load(f)
    
with open('acc_dropout_020.txt', 'rb') as f:
    acc_dropout_020 = pickle.load(f)
with open('acc_dropout_020_validation.txt', 'rb') as f:
    acc_dropout_020_validation = pickle.load(f)
```

```python
plt.figure(figsize=(10,5), dpi=100)
plt.xlabel("epoch")
plt.ylabel("training accuracy")
plt.plot([i for i in range(100)], acc_no_dropout, label="no dropout", linestyle='--')
plt.plot([i for i in range(100)], acc_dropout_005, label="dropout = 0.05", linestyle='-')
plt.plot([i for i in range(100)], acc_dropout_010, label="dropout = 0.10", linestyle=':')
plt.plot([i for i in range(100)], acc_dropout_015, label="dropout = 0.15", linestyle='-.')
plt.plot([i for i in range(100)], acc_dropout_020, label="dropout = 0.20", linestyle=':')
plt.legend(loc='lower right')
plt.savefig("training accuracy of different dropout")


plt.subplot()
plt.figure(figsize=(10,5), dpi=100)
plt.xlabel("epoch")
plt.ylabel("validation accuracy")
plt.plot([i for i in range(100)], acc_no_dropout_validation, label="no dropout", linestyle='-.')
plt.plot([i for i in range(100)], acc_dropout_005_validation, label="dropout = 0.05", linestyle='-')
plt.plot([i for i in range(100)], acc_dropout_010_validation, label="dropout = 0.10", linestyle=':')
plt.plot([i for i in range(100)], acc_dropout_015_validation, label="dropout = 0.15", linestyle=':')
plt.plot([i for i in range(100)], acc_dropout_020_validation, label="dropout = 0.20", linestyle='--')
plt.legend(loc='lower right')
plt.savefig("validation accuracy of different dropout")
```

```python
plt.figure(figsize=(10,5), dpi=100)
plt.xlabel("epoch")
plt.ylabel("training accuracy")
plt.plot([i for i in range(100)], dense8, label="unit = 8", linestyle=':')
plt.plot([i for i in range(100)], dense16, label="unit = 16", linestyle=':')
plt.plot([i for i in range(100)], dense32, label="unit = 32", linestyle='-')
plt.plot([i for i in range(100)], dense48, label="unit = 48", linestyle='-.')
plt.plot([i for i in range(100)], dense64, label="unit = 64", linestyle='-.')
plt.plot([i for i in range(100)], dense80, label="unit = 80", linestyle='-.')
plt.legend(loc='lower right')
plt.savefig("training accuracy of different unit")

plt.subplot()
plt.figure(figsize=(10,5), dpi=100)
plt.xlabel("epoch")
plt.ylabel("validation accuracy")
plt.plot([i for i in range(100)], dense8_validation, label="unit = 8", linestyle=':')
plt.plot([i for i in range(100)], dense16_validation, label="unit = 16", linestyle=':')
plt.plot([i for i in range(100)], dense32_validation, label="unit = 32", linestyle='-')
plt.plot([i for i in range(100)], dense48_validation, label="unit = 48", linestyle='-.')
plt.plot([i for i in range(100)], dense64_validation, label="unit = 64", linestyle='-.')
plt.plot([i for i in range(100)], dense80_validation, label="unit = 80", linestyle='-.')
plt.legend(loc='lower right')
plt.savefig("validation accuracy of different unit")
```
